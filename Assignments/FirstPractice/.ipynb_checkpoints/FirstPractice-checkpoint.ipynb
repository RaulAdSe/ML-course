In order to give a qualitative shape of the dependency of numerical variables with price we are going to explicitly describe the function type and let the scipy optimizer fit our parameters so the error is minimized. This will be done only with a pair of examples and the others will be done by hand. 

- For the first dependency an exponential delay or a power law could explain well the reltionship. Makes sense that very expensive cars, like supercars, have a much higher city consumption. This also matches with histograms of city consumption and price. There are very few cars that consume a lot and that are very expensive while the majority of them are affordable and have a mid-range (around 20-25 miles/gallon) city consumption.
- The second one is clearly linear. The bigger the engine size, the more expensive the car is. This could have been expected since the histogram of engine size and price has the same shape. Linear relationships are characterized by histograms of both variables having the same shape.

By analysis of the plots, one can tell the data of city-mpg is not ordered. In order to do a fitting it is cruciat for it to be.

```python
import matplotlib.pyplot as plt
import numpy as np

xdata = Data[' city-mpg']
ydata = Data[' price']

# delete all nans values
nan_indices_1 = np.where(np.isnan(xdata))
nan_indices_2 = np.where(np.isnan(ydata))

# use union1d to join the arrays without repetition
joined_arr = np.union1d(nan_indices_1, nan_indices_2)

xdata = np.delete(np.array(xdata), joined_arr)
ydata = np.delete(np.array(ydata), joined_arr)

# solve index problems from deleting rows previously by reindexing
if len(joined_arr)!= 0:
    new_index = range(len(ydata))
    xdata = xdata.reindex(new_index)
    ydata = ydata.reindex(new_index)

# get indices that would sort x in ascending order
sort_indices = np.argsort(xdata)

# sort both x and y using the same indices
x_sorted = xdata[sort_indices]
y_sorted = ydata[sort_indices]
```

```python
from scipy.optimize import curve_fit
from scipy.optimize import OptimizeWarning
import warnings


def func_a(x, a, b):
    return a * x ** -b

def func_b(x, a, b, c):
    return a * np.exp(-b * (x - c))


xdata = Data[' city-mpg']
ydata = Data[' price']

with warnings.catch_warnings():
    warnings.simplefilter('ignore', OptimizeWarning)
    popt_a, pcov = curve_fit(func_a, xdata, ydata)
    popt_b, pcov = curve_fit(func_b, xdata, ydata, p0 = [2.98639842e+04, 9.67477725e-02,15])
Data.plot.scatter(figsize=(10,6),y=' price',x=' city-mpg', label='Data')
plt.plot(x_sorted, func_a(x_sorted, *popt_a), 'g-', label='Fit_power')
plt.plot(x_sorted, func_b(x_sorted, *popt_b), 'r-', label='Fit_exponential')
plt.legend()
plt.show()
```

The preprocessing of the data is not as important as the last case when trying to fit due to the simplicity of a linear fitting.


```python
def func_b(x, a, b):
    return a + b * x
xdata = Data[' engine-size']
ydata = Data[' price']
popt, pcov = curve_fit(func_b, xdata, ydata)
Data.plot.scatter(figsize=(10,6),y=' price',x=' engine-size', label='Data')
plt.plot(xdata, func_b(xdata, *popt), 'r-', label='Fit')
plt.legend()
plt.show()
```

```python
# define target feature
target_feature = ' price'

# loop over all numerical features (excluding the target feature)
numerical_cols = Data.select_dtypes(include=['float64', 'int64']).columns.tolist()

# create a figure and axis object
fig, axs = plt.subplots(nrows=len(numerical_cols)-1, ncols=2, figsize=(10, 40)) #-1 since no need to plot price vs price

i = 0
for feature in (numerical_cols):
    if feature != target_feature:
        # plot the x-y scatter plot
        axs[i, 0].scatter(Data[feature], Data[target_feature])
        axs[i, 0].set_xlabel(feature)
        axs[i, 0].set_ylabel(target_feature)

        # plot the histogram of the feature
        axs[i, 1].hist(Data[feature], bins=20)
        axs[i, 1].set_xlabel(feature)
        i += 1
        
plt.tight_layout()
plt.show()
```

From this dependencis we can now write an intuitive and first approach y=f(x) wich relates features x with target y, price. 
- Let f(Weel base) be the fitted to price quadratic expression.
- Let g(length) be the fitted to price exponential expression. 
- Let h(width) be the fitted to price quadratic expression. 
- Let k(curb weight) be the fitted to price quadratic expression. 
- Let l(engine size) be the fitted to price linear expression. 
- Let t(horsepower) be the fitted to price quadratic expression. 
- Let q(city mpg) be the fitted to price exponential expression. 
- Let p(highway mpg) be the fitted to price exponential expression. 

A first approach to y would be to make a ponderated sum of all this expressions. $w_i$ be the weights of this ponderation. This weights $w_i$ should be normalized and would depend on the importance of the feature over the target. Thus:
$$
y = w_1*f(Weel base)+w_2*g(length)+w_3*h(width)+w_4*k(curb weight)+w_5*l(engine size)+w_6*t(horsepower)+w_7*q(city mpg)+w_8*p(highway mpg)
$$


A first approach to these wegihts would be the normalization of the absolute values of the correlation matrix.

Since the target is discrete, these kind of plots don't help us to state the shape of the dependency. We will need barplots where we can see the effects of one variables over the target.

We will use the barplot of seaborn package, which has the following description in its documentation. As in a classification problem, we will plot our target variable in the x axis.
Show point estimates and errors as rectangular bars.
A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars. Bar plots include 0 in the quantitative axis range, and they are a good choice when 0 is a meaningful value for the quantitative variable, and you want to make comparisons against it.

```python
# define target feature
target_feature = 'quality'

# loop over all numerical features (excluding the target feature)
numerical_cols = Data.select_dtypes(include=['float64', 'int64']).columns.tolist()

# create a figure and axis object
fig, axs = plt.subplots(nrows=len(numerical_cols)-1, ncols=2, figsize=(10, 40)) #-1 since no need to plot price vs price
warnings.filterwarnings('ignore', message='Pass the following variables as keyword args: x, y.*')
i = 0
for feature in (numerical_cols):
    if feature != target_feature:
        # plot the x-y scatter plot
        sns.barplot(Data[target_feature],Data[feature],ax = axs[i, 0], palette="GnBu_d")
        axs[i, 0].set_xlabel(target_feature)
        axs[i, 0].set_ylabel(feature)

        # plot the histogram of the feature
        axs[i, 1].hist(Data[feature], bins=20)
        axs[i, 1].set_xlabel(feature)
        i += 1
        
plt.tight_layout()
plt.show()
```

16. fuel-system: Mechanical Feature  (a factor).
- 1bbl, 2bbl, 4bbl, idi, mfi, mpfi, spdi, spfi.
17. bore: Mechanical Feature  (numeric).	
- continuous from 2.54 to 3.94.
18. stroke: Mechanical Feature  (numeric).

19. compression-ratio: Mechanical Feature  (numeric).	  
- continuous from 7 to 23.
20. horsepower: The power of the engine in horsepowers (numeric).	
- continuous from 48 to 288.
21. peak-rpm: The top speed of the engine in rounds-per-minute (numeric).
- continuous from 4150 to 6600.
22. city-mpg: The fuel consumption of the car in city driving conditions, measured as miles per gallon of fuel (numeric).
