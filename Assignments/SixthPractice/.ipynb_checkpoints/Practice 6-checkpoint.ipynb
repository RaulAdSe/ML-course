## Let us train a classifier

##  First we partition database into train and test
### Note the helper function

```python
def PartitionOfDatabase(Data,Fraction = 0.75):
    NumSamples, NumFeatures = Data.shape
    Index = list(range(NumSamples))
    RandomIndex = np.random.permutation(Index)
    NumTrainElements = int(Fraction*NumSamples)
    TrainIndex = RandomIndex[:NumTrainElements]
    TestIndex = RandomIndex[NumTrainElements:]
    TrainDB = Data.iloc[TrainIndex]
    TestDB = Data.iloc[TestIndex]
    return TrainDB,TestDB
TrainDB,TestDB = PartitionOfDatabase(Data,Fraction = 0.75)
```

```python
y = TrainDB['chd']
InputFeatures = [fea for fea in TrainDB.columns if  fea != 'chd']
print(f' Input variables: \n {InputFeatures}')
print('_'*50)
X = TrainDB[InputFeatures]
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Calculate the Gini coefficient for each node in the decision tree
fig, ax = plt.subplots(figsize=(20,10))
plot_tree(clf, filled=True, impurity=True, feature_names=InputFeatures, ax=ax)

# Add a colorbar to the plot
impurity_values = clf.tree_.impurity
norm = plt.Normalize(vmin=impurity_values.min(), vmax=impurity_values.max())
cmap = plt.cm.RdYlBu  # choose a colormap
sm = plt.cm.ScalarMappable(norm=norm, cmap=cmap)
sm.set_array([])  # this line is necessary for the colorbar to work properly
cbar = plt.colorbar(sm)
cbar.set_label('Impurity')

plt.show()
```

The Gini coefficient measures the impurity of a node in a decision tree, with values ranging from 0 (perfectly pure) to 0.5 (equally impure). The closer the Gini coefficient is to 0, the more homogenous the class distribution is in that node.

In a decision tree, the root node represents the entire dataset, and as such, it is expected to have a higher Gini coefficient compared to the lower nodes that represent smaller subsets of the data. As the tree splits into lower nodes, the subsets become smaller and more homogenous, resulting in a lower Gini coefficient.

Regarding the variable "Value," it gives the number of elements of each class in a node. The number of elements in each class affects the Gini coefficient, as a node with a higher number of elements in one class will have a higher Gini coefficient if the other class is represented by a much smaller number of elements.

## Let us control the training of the tree

1. min_samples_split: We want to control the number in order to have significative statistics.

2. max_depth: depth we allow for the tree. Will stop growth.

3. min_samples_leaf: We want to have enoughs examples per decision.

4. class_weight:  Allows to control balance between *True Positives* and *False Negatives*.

```python
clf = tree.DecisionTreeClassifier(min_samples_split = 10,max_depth =5,min_samples_leaf =15,class_weight = {0:0.5,1:0.5})
clf = clf.fit(X, y)
plt.figure(figsize=(25,25))
ax = tree.plot_tree(clf,feature_names = InputFeatures,impurity = True,class_names =['no','chd'],filled = True)
plt.show()
```

When we set class_weight to a value that assigns higher weight to the minority class, the model pays more attention to correctly classifying the minority class during the training process. This is because the cost of misclassifying the minority class is higher than that of misclassifying the majority class.

Mathematically, the class_weight parameter affects the calculation of the splitting criterion used by the decision tree algorithm. The default splitting criterion in scikit-learn is the Gini impurity, which measures the probability of misclassifying a sample at a given node. When class_weight is set, the Gini impurity is modified to take into account the class weights. Specifically, the impurity at a node is calculated as:

$\sum_{i=1}^{k} w_i (1 - p_i^2)$

where k is the number of classes, $w_i$ is the weight assigned to class i, and $p_i$ is the proportion of samples in class i at the node. By default, all weights are set to 1, so class_weight = None is equivalent to class_weight = {0:1,1:1}. When we set class_weight to a value that assigns higher weight to the minority class, the Gini impurity will be modified accordingly, resulting in a tree that is biased towards correctly classifying the minority class.

Consequently, we observe an increasy in accuracy in both databases, train and test.

### Homework: <font color='red'>   Try a few changes in min_samples_split = 10,max_depth =5,min_samples_leaf =15.</font> 

  <font color='red'>  Note what happen when you make it more strict or more loose. </font>
  
 <font color='red'>   **Contrast** the effect on the training performance and test performance. Remember that the important performance is on **test** </font>

Let's make some code for hyperpameter tunning.

```python
best_params = None
best_test_acc = 0

# initialize the classifier with default values
clf = DecisionTreeClassifier(random_state=42)

# fit the model on the training data
clf.fit(X_train, y_train)

# predict on the training and test data
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)

# calculate accuracy on training and test data
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print(f"Training accuracy: {train_acc:.2f}")
print(f"Test accuracy: {test_acc:.2f}")

# try out different values of hyperparameters
test_accs = []
for min_samples_split in [5, 10, 20]:
    for max_depth in [3, 5, 7]:
        for min_samples_leaf in [5, 10, 15]:
            # initialize the classifier with different hyperparameters
            clf = DecisionTreeClassifier(min_samples_split=min_samples_split, 
                                          max_depth=max_depth, 
                                          min_samples_leaf=min_samples_leaf)
            # fit the model on the training data
            clf.fit(X_train, y_train)
            
            # predict on the training and test data
            y_train_pred = clf.predict(X_train)
            y_test_pred = clf.predict(X_test)
            
            # calculate accuracy on training and test data
            train_acc = accuracy_score(y_train, y_train_pred)
            test_acc = accuracy_score(y_test, y_test_pred)
            test_accs.append(test_acc)
            
            # check if this is the best combination of hyperparameters
            if test_acc > best_test_acc:
                best_test_acc = test_acc
                best_params = {'min_samples_split': min_samples_split,
                               'max_depth': max_depth,
                               'min_samples_leaf': min_samples_leaf}

            print(f"min_samples_split: {min_samples_split}, max_depth: {max_depth}, min_samples_leaf: {min_samples_leaf}")
            print(f"Training accuracy: {train_acc:.3f}")
            print(f"Test accuracy: {test_acc:.3f}")
            print("")

print(f"Best combination of hyperparameters: {best_params}")
print(f"Test accuracy: {best_test_acc:.3f}")
```

In this code, we first initialize the DecisionTreeClassifier with default hyperparameters and fit the model on the training data. We then predict on the training and test data and calculate the accuracy on both sets.

We then try out different values of min_samples_split, max_depth, and min_samples_leaf using nested for-loops. For each combination of hyperparameters, we initialize a new classifier and fit the model on the training data. We then predict on the training and test data and calculate the accuracy on both sets.

By varying the hyperparameters, we can make the model more or less strict. A smaller min_samples_split means the tree will split a node if it has at least that many samples, which could lead to overfitting. A larger max_depth means the tree will go deeper, potentially leading to overfitting as well. A smaller min_samples_leaf means each leaf node must have at least that many samples, which could result in more complex trees.

By comparing the training and test accuracies for different hyperparameters, we can find the optimal values that generalize well to unseen data. I have chosen the best combination of hyperparameters in a naive way, however, as one may see, different combination of parameters yield the same test accuracy. I have highlighted only one of the best combination of hyperparameters.
