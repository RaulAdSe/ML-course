## Partition database into train and test

Note the helper function

```python
def PartitionOfDatabase(Data,Fraction = 0.75):
    NumSamples, NumFeatures = Data.shape
    Index = list(range(NumSamples))
    RandomIndex = np.random.permutation(Index)
    NumTrainElements = int(Fraction*NumSamples)
    TrainIndex = RandomIndex[:NumTrainElements]
    TestIndex = RandomIndex[NumTrainElements:]
    TrainDB = Data.iloc[TrainIndex]
    TestDB = Data.iloc[TestIndex]
    return TrainDB,TestDB
TrainDB,TestDB = PartitionOfDatabase(Data,Fraction = 0.75)
```

## We create the projection into 2D by the two methods.

The dimension of the projection in LDA is equal to the number of classes minus one, or min(C-1, P), where C is the number of classes and P is the number of input features (i.e. the dimension of the input space). The projection is a linear combination of the input features, where the coefficients of the linear combination are chosen to maximize the separability between the classes while minimizing the within-class variance. The resulting projection can be used to reduce the dimensionality of the input space while preserving the class-discriminatory information.

For example, if the original data had 11 features, LDA could project the data onto a 2-dimensional subspace to better separate the classes. In this subspace, the algorithm would create a decision boundary that maximizes the difference between the means of the classes while minimizing the variance within each class.

The dimension of projection is typically chosen based on a trade-off between the complexity of the model and its ability to accurately separate the classes. A lower-dimensional subspace may not capture all the information in the original data, while a higher-dimensional subspace may overfit to noise in the data.

To find the projection of our data in the lower-dimensional space we use the transform method of the trained LDA model.

```python
Dimension_of_projection = 2
```

```python
lda = LinearDiscriminantAnalysis(n_components=Dimension_of_projection)
lda_2 = LinearDiscriminantAnalysis(n_components=Dimension_of_projection)

X_lda = lda.fit(X, y).transform(X)
X_lda = pd.DataFrame(data = X_lda, columns = ['dim1','dim2'])

lda_2.fit(X_untransformed, y)
```

LDA finds a linear combination of features that maximizes the ratio of the between-class variance to the within-class variance. This is done by first computing the mean vector and covariance matrix for each class. Then, the within-class scatter matrix and between-class scatter matrix are calculated. The within-class scatter matrix measures the scatter of the data points within each class, while the between-class scatter matrix measures the scatter of the class means.

Next, LDA performs an eigenvalue decomposition of the matrix inverse of the within-class scatter matrix multiplied by the between-class scatter matrix. The eigenvectors with the highest absolute value eigenvalues are selected as the new dimensions, in this case the eigenvectors with the 2 highest eigenvalues. The projection of the data onto these dimensions forms the new feature space that maximizes the separation between the classes.

```python
lda.scalings_
```

This is an array of shape (n_features, n_components), where n_features is the number of features in your data and n_components is the number of dimensions in your projection (in your case, n_components is 2). Each column of this array represents a dimension of the projection, and each row represents a feature in your data. The values in the array are the coefficients of each feature in the corresponding dimension of the projection.

## Homework: <font color='red'> State briefly the criterion for deriving each method and the underlying hypothesis.</font>

Linear discriminant analysis (LDA) is a statistical method used to find a linear combination of features that can best discriminate between two or more classes or groups. The criterion for deriving LDA is to maximize the between-class variance while minimizing the within-class variance of the data. The underlying hypothesis is that the data points in each class have a multivariate normal distribution with a shared covariance matrix and that the mean vectors of the classes are different from each other.


### Homework: <font color='red'>  Given a multidimensional Gaussian with $x \in \mathbb{R}^p$, which is the distribution of a linear combination of the dimensions of $x$ ?</font>

If we have a multidimensional Gaussian distribution with $x \in \mathbb{R}^p$, where  is a p-dimensional random vector with mean  and covariance matrix , and we want to find the distribution of a linear combination of the dimensions of , then the resulting distribution will also be Gaussian.

Say $x \sim \mathcal{N}_p(\boldsymbol{\mu}, \mathbf{\Sigma})$, then a linear combination of $x$ would be:
$y = a' x$, where a is a vector in $\mathbb{R}^p$.

It would have the following expected value: $E(y) = E(a' x) = E(a') E(x) = a' $

It would have the following variance: $Var(y) = a' Var(x) a = a'  a$

Therfore, $y \sim \mathcal{N}\left(a^{\prime} \boldsymbol{\mu}, a^{\prime} \boldsymbol{\Sigma} a\right)$.

Despite their shape and the above reasoning we could consider the above distributions as approximately normal. Performing statistical tests for normality such as the Shapiro-Wilk test or the Anderson-Darling test would give us a binary YES/NO answer.



```python
from scipy import stats

# Perform Shapiro-Wilk test
stat, p = stats.shapiro(X_lda['dim1'])

# Print test statistics and p-value
print('Test statistics:', stat)
print('p-value:', p)

# Interpret the result
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian for dim1 (reject H0)')
    
# Perform Shapiro-Wilk test
stat, p = stats.shapiro(X_lda['dim2'])

# Print test statistics and p-value
print('Test statistics:', stat)
print('p-value:', p)

# Interpret the result
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian for dim2 (reject H0)')
```

We observe the statistical test yields a negative response. However, we will consider them (more or less) Gaussian distributions. If the original source of this linear combinations would be perfectly Gaussian, we would have obtained a more Gaussian shape.

In LDA, the goal is to find a linear combination of the features that best separates the different classes. The coefficients in LDA represent the weights assigned to each feature in this linear combination. Specifically, the coefficients are derived by maximizing a criterion that measures the separation between the classes, while also minimizing the variance within each class.

Each class has different coefficients in LDA because the weights assigned to each feature depend on the mean and variance of that feature within each class. The coefficients are calculated separately for each class in order to maximize the separation between the classes. In other words, the coefficients for each class are optimized to create a linear boundary that best separates that class from the other classes in the dataset.

```python
Coef_Lda_untransformed = pd.DataFrame(data = lda_2.coef_.T, columns = ['Proj class 1','Proj class 2','Proj class 3'],index =X.columns.values)
Coef_Lda = pd.DataFrame(data = lda.coef_.T, columns = ['Proj class 1','Proj class 2','Proj class 3'],index =X.columns.values)
Coef_Lda
```

The coefficients for each feature of each class represent the contribution of that feature to the discriminant function that separates the different classes. The coefficients are chosen in such a way that the linear combinations of the original features maximize the separation between the means of the classes while minimizing the variance within each class. The resulting linear combinations of features are called discriminant functions, and they can be used to classify new data points into one of the classes.

However one must be aware that discriminant functions and classification boundaries are related concepts, but they are not exactly the same.

Discriminant functions are mathematical functions used in LDA to project data onto a lower-dimensional space while maximizing the class separation. They are used to calculate the linear combination of the input features that best separates the classes.  Each discriminant function is a linear combination of the input features and has a set of coefficients that are calculated during the training process.

Classification boundaries, on the other hand, are decision boundaries that separate the classes in a classification problem. They can be linear or non-linear and are determined by the discriminant functions. In a two-class problem, a linear discriminant function can be used to find a linear classification boundary that separates the two classes. The boundary is defined as the set of points where the discriminant function is equal to a threshold value. Points on one side of the boundary are classified as belonging to one class, while points on the other side are classified as belonging to the other class.

In an attempt to make a clarification of the above, we will use the linear transformation of data that the scalings_ functions give us in order to draw the decision boundaries.

```python
# Create a meshgrid of points in the reduced space to plot the decision boundaries
x_min, x_max = X_lda['dim1'].min() - 1, X_lda['dim1'].max() + 1
y_min, y_max = X_lda['dim2'].min() - 1, X_lda['dim2'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))

# Get the matrix used to reduce the data during training
W = lda.scalings_

# Multiply the reduced coordinates by the transpose of W to get the full size coordinates
full_coords = np.dot(np.c_[xx.ravel(), yy.ravel()], W.T)

X_grid_full_size = full_coords[:,:len(full_coords)-1]

# Use the full size transformed grid coordinates to make predictions
Z = lda.predict(X_grid_full_size)

# Reshape Z and plot the decision boundaries as before
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.4)

# Plot the transformed data and decision boundaries
plt.scatter(X_lda['dim1'], X_lda['dim2'], c=y, cmap='viridis')
plt.xlabel('dim1')
plt.ylabel('dim2')
plt.title('Decision boundaries projected into the reduced space')
```

From a theoretical point of view, the coefficients in linear discriminant analysis (LDA) represent the weights assigned to each feature in a linear combination used to separate the different classes. These coefficients are calculated in such a way that the linear combination maximizes the separation between the classes, based on the within-class and between-class variance of the features.

In practice, the sign and size of the coefficients can be used to interpret the relationship between each feature and the classification task at hand. For example, a positive coefficient for a particular feature indicates that higher values of that feature tend to be associated with one class, while a negative coefficient indicates the opposite. The magnitude of the coefficient indicates the relative importance of the feature in the classification task, with larger magnitudes indicating greater importance.

We first do it in the original space.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.utils import shuffle

# Split data into three groups based on class labels
class_1_train = TrainDB[TrainDB['quality'] == 1][InputFeatures]
class_2_train = TrainDB[TrainDB['quality'] == 2][InputFeatures]
class_3_train = TrainDB[TrainDB['quality'] == 3][InputFeatures]

class_1_test = TestDB[TestDB['quality'] == 1][InputFeatures]
class_2_test = TestDB[TestDB['quality'] == 2][InputFeatures]
class_3_test = TestDB[TestDB['quality'] == 3][InputFeatures]

y_trains = [[np.hstack((np.ones(len(class_1_train)), np.zeros(len(class_2_train)), np.zeros(len(class_3_train))))],
           [np.hstack((np.zeros(len(class_1_train)), np.ones(len(class_2_train)), np.zeros(len(class_3_train))))],
           [np.hstack((np.zeros(len(class_1_train)), np.zeros(len(class_2_train)), np.ones(len(class_3_train))))]]

y_tests = [[np.hstack((np.ones(len(class_1_test)), np.zeros(len(class_2_test)), np.zeros(len(class_3_test))))],
           [np.hstack((np.zeros(len(class_1_test)), np.ones(len(class_2_test)), np.zeros(len(class_3_test))))],
           [np.hstack((np.zeros(len(class_1_test)), np.zeros(len(class_2_test)), np.ones(len(class_3_test))))]]


X_train = np.vstack((class_1_train,class_2_train, class_3_train))
X_test = np.vstack((class_1_test,class_2_test, class_3_test))

# Loop through three logistic models
for i in range(3):
    
    # Define logistic regression models
    logreg = LogisticRegression(solver='newton-cg',penalty = 'l2',C=10,max_iter = 1000,fit_intercept = True)
    
    y_train = y_trains[i][0]
    y_test = y_tests[i][0]

    # Shuffle the data 
    X_train, y_train = shuffle(X_train, y_train)
    logreg.fit(X_train, y_train)

    X_test, y_test = shuffle(X_test, y_test)

    # predict on the testing data
    y_pred = logreg.predict(X_test)

    # evaluate the model's performance
    accuracy_test = accuracy_score(y_test, y_pred)

    y_pred_train = logreg.predict(X_train)
    accuracy_train = accuracy_score(y_train, y_pred_train)

    print("{}th logistic classifier yields {} in test data and {} in train data.".format(i+1, accuracy_test, accuracy_train))
    
    # plot the coefficients
    fig, ax = plt.subplots(figsize=(25, 4))
    ax.bar(np.arange(logreg.coef_.shape[1]), logreg.coef_[0])
    ax.set_xticks(np.arange(logreg.coef_.shape[1]))
    ax.set_xticklabels(InputFeatures)
    ax.set_ylabel('Coefficient')
    plt.show()
```

When a logistic regression classifier has the same accuracy on test data as it does on train data, it means that the model is not overfitting or underfitting the data. In other words, the model is able to generalize well and make accurate predictions on unseen data. This is an indication that the model has learned the underlying patterns in the data and can make predictions on new data points.

```python
lda = LinearDiscriminantAnalysis(n_components=Dimension_of_projection)

X_lda = lda.fit(X_train, y_train).transform(X_train)
X_lda = pd.DataFrame(data = X_lda, columns = ['dim1','dim2'])

# Create a meshgrid of points in the reduced space to plot the decision boundaries
x_min, x_max = X_lda['dim1'].min() - 1, X_lda['dim1'].max() + 1
y_min, y_max = X_lda['dim2'].min() - 1, X_lda['dim2'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                 	np.arange(y_min, y_max, 0.05))

# Get the matrix used to reduce the data during training
W = lda.scalings_

# Multiply the reduced coordinates by the transpose of W to get the full size coordinates
full_coords = np.dot(np.c_[xx.ravel(), yy.ravel()], W.T)

X_grid_full_size = full_coords[:,:len(full_coords)-1]

# Use the full size transformed grid coordinates to make predictions
Z = lda.predict(X_grid_full_size)

# Reshape Z and plot the decision boundaries as before
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap='viridis', alpha=0.4)

# Plot the transformed data and decision boundaries
plt.scatter(X_lda['dim1'], X_lda['dim2'], c=y, cmap='viridis')
plt.xlabel('dim1')
plt.ylabel('dim2')
plt.title('Decision boundaries projected into the reduced space')

Means_Lda = pd.DataFrame(data = lda.means_.T, columns = ['Proj 1','Proj 2','Proj 3'],index =X.columns.values)

ax =Means_Lda.plot.bar()

# Plot centroids of all features
from mpl_toolkits.mplot3d import Axes3D

# assume means is a numpy array with shape (num_classes, 3)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# plot the centroids as red circles
ax.scatter(Means_Lda['Proj 1'], Means_Lda['Proj 2'], Means_Lda['Proj 3'], c='r', marker='o')

# set the axis labels
ax.set_xlabel('Proj 1')
ax.set_ylabel('Proj 2')
ax.set_zlabel('Proj 3')
plt.title('Centroids of the three classes')
plt.show()
```

In the cofficient barplot we see how the new features added in the feature engineering play a role. The centroids seem a little bit more spaced than with original data (before feature engineering), which will lead to a better classification.
